```
y = dataFrame["price"].values # We've equalized the price to y and converted it into a numpy array.
x = dataFrame.drop("price",axis=1).values # We've removed the "price" column from the "dataFrame" and equated it to x.
```

```
from sklearn.model_selection import train_test_split # We imported "train_test_split" from "sklearn.model_selection"."train_test_split" will help us to divide our data into " test "and" train".
```

```
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.30,random_state=10) # We will use "_train" for train our model and "_test" for test it.
```

```
len(x_train) # We've looked at length of "x_train."
```

```
len(x_test) # We looked at length of "x_test".
```

```
 from sklearn.preprocessing import MinMaxScaler # We imported "MinMaxScaler" from "sklearn.preprocessing". We use "MinMaxScaler" for scale x value.
```

```
scaler = MinMaxScaler() # We created new object from " MinMaxScaler."
```

```
x_train = scaler.fit_transform(x_train) # We fitted the "x_train" value to between 0 and 1.
```

```
x_test = scaler.transform(x_test) 
```

```
from tensorflow.keras.models import Sequential # The class which is we've created model.
from tensorflow.keras.layers import Dense # The class whic is we've created layers.
```

```
x_train.shape
```

```
model = Sequential()

model.add(Dense(12,activation="relu")) # We gave it 12 neurons.
model.add(Dense(12,activation="relu")) # We gave it 12 neurons.
model.add(Dense(12,activation="relu")) # We gave it 12 neurons.
model.add(Dense(12,activation="relu")) # We gave it 12 neurons.
model.add(Dense(12,activation="relu")) # We gave it 12 neurons.
                                       # We did this 5 times which is mean we have 5 layers.
                                       # We've choosen "relu" for "Activation".
                                       
model.add(Dense(1)) # We gave 1 output layer.

model.compile(optimizer="adam",loss="mse") # "adam" gives us best results because of this we use "adam". "loss" gives us losses. How much closer "loss" value to 0 this means we trained our model well.
#If you give su much neurons or layers it can be overfitting. Overfitting is "the production of an analysis that corresponds too closely or exactly to a particular set of data, and may therefore fail to fit additional data or predict future observations reliably".
```

```
model.fit(x = x_train, y = y_train,validation_data=(x_test,y_test),batch_size=250,epochs=400) # We've synchronized our train data to x and y.
# "validation_data=" allows us to benchmarking process between "x_test" and "y_test" on the same screen.
# "batch_size" allows us to give data part by part.Because if we give 13000 data at the same time, we tire the model.For example "batch_size=250" means it gives 250 datas at 1 time.
```

```
lostData = pd.DataFrame(model.history.history) # We've defined the resulting result in a new variable that we create with the name "lostData". You can use different name of course.
```
